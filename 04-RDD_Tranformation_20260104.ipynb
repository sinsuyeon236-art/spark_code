```python
# 2.1. Narrow Transformations
```

```python
# parallelize()
sample_rdd =  sc.parallelize([1,2,3])
sample_rdd
```

```python
sample_rdd2 = sample_rdd.map(lambda x : x + 2)
sample_rdd2
```

```python
sample_rdd2.collect()
```

```python
# flatMap()
movies = [
    "그린 북",
    "매트릭스",
    "토이 스토리",
    "캐스트 어웨이",
    "포드 V 페라리",
    "보헤미안 랩소디",
    "빽 투 더 퓨처",
    "반지의 제왕",
    "죽은 시인의 사회"
]
```

```python
moviesRDD = sc.parallelize(movies)
moviesRDD
```

```python
mapMovies = moviesRDD.map(lambda x : x.split())
mapMovies.collect()
```

```python
flatMovies = moviesRDD.flatMap(lambda x : x.split())
flatMovies.collect()
```

```python
# 2.2. Wide Transformations
```

```python
foods = sc.parallelize([
    "짜장면", "마라탕", "짬뽕", "떡볶이", "쌀국수", "짬뽕", "짜장면", "짜장면", "짜장면", "라면", "우동", "라면", "라볶이", "짜장밥", "짬짜면"
])
foods
```

```python
# 첫 글자 기준
foodsGroup = foods.groupBy(lambda x : x[0])
foodsGroup
```

```python
result = foodsGroup.collect()
result
```

```python
for (k,v) in result:
    print(k, list(v))
```
