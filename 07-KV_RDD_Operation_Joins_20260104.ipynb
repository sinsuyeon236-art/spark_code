```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("key-value_rdd_op_joins")
sc = SparkContext(conf=conf)
```

# Operations
- `groupByKey`
    - `KeyValueRDD.groupByKey()`
    - 그루핑 후에 특정 Transformations 연산
    - `key`값이 존재하는 상태로 시작

```python
rdd = sc.parallelize([
    ("짜장면", 15),
    ("짬뽕", 10),
    ("짜장면", 5)
])
```

```python
g_rdd = rdd.groupByKey()
g_rdd
```

```python
g_rdd.collect()
```

```python
# len : 파이썬에서 리스트나 튜플 등의 컬렉션 길이를 반환하는 함수
g_rdd.mapValues(len).collect()
```

```python
g_rdd.mapValues(list).collect()
```

# Joins

```python
rdd1 = sc.parallelize([
    ("foo", 1),
    ("goo", 2),
    ("hoo", 3)
])
```

```python
rdd2 = sc.parallelize([
    ("foo", 1),
    ("goo", 2),
    ("goo", 10),
    ("moo", 6),
])
```

```python
# Inner Join : 서로 간에 존재하는 키만 합쳐준다.
rdd1.join(rdd2).collect()
```

```python
rdd1.leftOuterJoin(rdd2).collect()
```

```python
rdd1.rightOuterJoin(rdd2).collect()
```

```python
rdd1.fullOuterJoin(rdd2).collect()
```
